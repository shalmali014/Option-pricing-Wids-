```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#for other we use plt.show()
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#exercise 1
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('x vs y')
```

```python
#Exercise 2
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y)
ax2.plot(x, y)
```

```python
#Exercise 3
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x, y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Main Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Inset Plot')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, y)
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
```

```python
#Exercise 4
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, linewidth=2, linestyle='--')
axes[0].set_title('x vs y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x, z, linewidth=3, linestyle='-')
axes[1].set_title('x vs z')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
axes[0].plot(x, y, linewidth=2, linestyle='--')
axes[0].set_title('x vs y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x, z, linewidth=3, linestyle='-')
axes[1].set_title('x vs z')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
