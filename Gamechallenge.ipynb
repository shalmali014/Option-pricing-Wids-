```python
import random

num = random.randint(1, 100)
print("Guess a number between 1 and 100")

guesses = [0]

```

```python
while True:
    guess = int(input("Enter your guess: "))
    if guess < 1 or guess > 100:
        print("OUT OF BOUNDS")
    else:
        guesses.append(guess)
        break

```

```python
while True:
    if guess == num:
        print(f"Correct! You guessed it in {len(guesses)-1} guesses.")
        break

    if len(guesses) == 2:
        if abs(num - guess) <= 10:
            print("WARM!")
        else:
            print("COLD!")
    else:
        if abs(num - guess) < abs(num - guesses[-2]):
            print("WARMER!")
        else:
            print("COLDER!")

    guess = int(input("Enter your guess: "))
    if guess < 1 or guess > 100:
        print("OUT OF BOUNDS")
        continue
    guesses.append(guess)
```
