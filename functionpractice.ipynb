```python
# Q1
def lesser_of_two_evens(a, b):
    if a % 2 == 0 and b % 2 == 0:
        return min(a, b)
    else:
        return max(a, b)
```

```python
lesser_of_two_evens(2,4)
```

```python
lesser_of_two_evens(2,5)
```

```python
# Q2
def animal_crackers(text):
    words = text.lower().split()
    return words[0][0] == words[1][0]
```

```python
# Q7
def has_33(nums):
    for i in range(len(nums) - 1):
        if nums[i] == 3 and nums[i + 1] == 3:
            return True
    return False
```

```python
# Check
has_33([1, 3, 3])
```

```python
# Check
has_33([1, 3, 1, 3])
```

```python
# Check
has_33([3, 1, 3])
```

```python
# Q8
def paper_doll(text):
    result = ""
    for char in text:
        result += char * 3
    return result
```

```python
# Check
paper_doll('Hello')
```

```python
# Check
paper_doll('Mississippi')
```

```python
# Q9
def blackjack(a, b, c):
    total = a + b + c
    if total <= 21:
        return total
    if total > 21 and 11 in [a, b, c]:
        total -= 10
    if total > 21:
        return "BUST"
    return total
```

```python
# Check
blackjack(5,6,7)
```

```python
# Check
blackjack(9,9,9)
```

```python
# Check
blackjack(9,9,11)
```

```python
# Q10
def summer_69(arr):
    total = 0
    ignore = False
    for num in arr:
        if num == 6:
            ignore = True
        elif num == 9 and ignore:
            ignore = False
        elif not ignore:
            total += num
    return total
```

```python
# Check
summer_69([1, 3, 5])
```

```python
# Check
summer_69([4, 5, 6, 7, 8, 9])
```

```python
# Check
summer_69([2, 1, 6, 9, 11])
```

```python
# Q11
def spy_game(nums):
    code = [0, 0, 7]
    for num in nums:
        if num == code[0]:
            code.pop(0)
        if not code:
            return True
    return False
```

```python
# Check
spy_game([1,2,4,0,0,7,5])
```

```python
# Check
spy_game([1,0,2,4,0,5,7])
```

```python
# Check
spy_game([1,7,2,0,4,5,0])
```

```python
# Q12
def count_primes(num):
    if num < 2:
        return 0
    primes = [2]
    for n in range(3, num + 1, 2):
        for p in primes:
            if n % p == 0:
                break
            if p * p > n:
                primes.append(n)
                break
    return len(primes)
```

```python
# Check
count_primes(100)
```
